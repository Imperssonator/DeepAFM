```python
the_csv_file = 'file:///Users/Imperssonator/CC/DeepAFM/Hu_imgs.csv'
dfpc = pd.read_csv(the_csv_file)
dfpc
```

```python
import pandas as pd

the_json_file = 'file:///Users/Imperssonator/CC/Sherlock/scrape_figures/output.json'
df = pd.read_json(path_or_buf=the_json_file)
df['journal_label']=df.groupby('journal').grouper.labels[0]
df['localName'] = [x[0]['path'][5:] for x in df['files']]
df = df.sort_values('localName')
df['localName'][576]
```

```python
df['localName'].isin([dfpc['name'][0]]).sum()
```

```python
name_matches = [df[df['localName']==x]['file_urls'] for x in df['localName'] if dfpc['name'].isin([x]).sum()==1]
dfpc['url'] = [x.iloc[0] for x in name_matches]
dfpc
```

```python
import time
import datetime
dt = datetime.datetime.strptime(df['pubDate'][0], '%Y-%m-%d')
df['pubDate']=df['pubDate'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
df['pubDate'].head()
```

```python
colors = {}
column_to_color = 'date'
for c in dfpc[column_to_color].unique():
    colors[c] = np.random.random_integers( 0,255, (1,3))
    dfpc.loc[dfpc[column_to_color]==c,'color'] = '#' + ''.join([ '%02x'%s for s in colors[c][0]] )
    
dfpc['color'].head()
```

```python
dfpc['url'][0]
```

```python
p=figure(responsive=False,x_axis_label="PC1",y_axis_label="PC2")

x_col = 'PC1'
y_col = 'PC2'

cds = bokeh.models.sources.ColumnDataSource(
    dfpc
)

circle = p.circle(
    source=cds,
    x=x_col,
    y=y_col,
    fill_color='color',
    size=20,
    fill_alpha=.6
)

# p.add_tools(
#     bokeh.models.HoverTool(
#         tooltips=[
#             ('title', '@title'),
#         ]
#     )
# )
#     h=0.003,
#     w=0.04,
images = p.image_url(
    source=cds,
    x=x_col,
    y=y_col,
    url='url',
    anchor='center',
    global_alpha=.001
)

p.x_range.callback = bokeh.models.CustomJS( args= {
        'x_axis': p.x_range,
        'y_axis': p.y_range,
        'images': images.glyph,
    }, 
    code=coffee.compile("""
    dx = x_axis.get('end') - x_axis.get('start')
    if dx > .1
        images.set 'global_alpha', .001 + (.8)*Math.exp( -1*Math.pow(dx-.1,2)/(0.014))
    """,bare=True)
)
```
